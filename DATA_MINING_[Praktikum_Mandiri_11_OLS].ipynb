```python
import pandas as pd
data = pd.read_csv('/content/Data Pelatihan 2024',sep = ',')
data
```

```python
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm

x = data[["Nilai", "UTS"]].to_numpy()
y = data["UAS"].to_numpy()

x = sm.add_constant(x)
model_prediksi = sm.OLS(y, x).fit()
print(model_prediksi.summary())
```

```python
data1 = pd.read_csv('/content/Data Pelatihan (testing)',sep = ',')
x_tes = data1[["Nilai", "UTS"]].to_numpy()
y_real = data1["UAS"].to_numpy()
y_nama = data1["Nama"].to_numpy()
x_tes = sm.add_constant(x_tes)
y_pred = model_prediksi.predict(x_tes)
```

```python
data = pd.DataFrame()
data["Nama Mahasiswa"]=y_nama
data["Nilai UAS Real"]=y_real
data["Nilai UAS Prediksi"]=y_pred
data["ERROR"]=data["Nilai UAS Real"]-data["Nilai UAS Prediksi"]
data["PERSEN PREDIKSI"]=1.0-(abs(data["ERROR"])/data["Nilai UAS Real"])
data["PERSEN"]=data["PERSEN PREDIKSI"]*100
display(data)
```

```python
data.to_csv('Data Hasil 2024.csv')
datasets = pd.read_csv("/content/Data Hasil 2024.csv")
datasets
```

```python
plt.figure(figsize=(12,6))
x = datasets.iloc[:, 0 ].values
plt.plot(x, datasets["Nilai UAS Real"], label = "Origin", color="b")
plt.plot(x, datasets["Nilai UAS Prediksi"], label = "Estimated", color="r")

plt.legend()
plt.xlabel("Simple ID")
plt.ylabel("Nilai UAS")
plt.title("Prediksi Nilai UAS Mahasiswa - OLS")
plt.show()
```
