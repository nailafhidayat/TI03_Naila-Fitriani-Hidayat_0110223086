```python
def calculate_outliers(df, column):
    # Hitung Q1, Q3, dan IQR
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1

    # Hitung batas bawah dan batas atas untuk outlier
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    # Temukan outlier dalam DataFrame
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return q1, q3, outliers

# Hitung outlier untuk kolom '2019'
q1_2019, q3_2019, outliers_2019 = calculate_outliers(df, '2019')
print("Q1 2019:", q1_2019)
print("Q3 2019:", q3_2019)
print("Outlier 2019:")
print(outliers_2019)

# Hitung outlier untuk kolom 'Total_import'
q1_Total_import, q3_Total_import, outliers_Total_import = calculate_outliers(df, 'Total_import')
print("Q1 Total_import:", q1_Total_import)
print("Q3 Total_import:", q3_Total_import)
print("Outlier Total_import:")
print(outliers_Total_import)
```
