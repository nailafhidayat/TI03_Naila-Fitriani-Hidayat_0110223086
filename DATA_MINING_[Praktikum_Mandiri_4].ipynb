```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# Load the dataset
df = pd.read_csv('/content/supermarket_sales - Sheet1.csv')
df.head()
```

```python
# Select the relevant columns for association rule mining
basket = df[['Product line', 'City']]
```

```python
# Create a one-hot encoded representation of the basket
basket_sets = basket.groupby(['City', 'Product line'])['Product line'].count().unstack().fillna(0).applymap(lambda x: 1 if x > 0 else 0)

# Apply the apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(basket_sets, min_support=0.05, use_colnames=True)

# Generate association rules from the frequent itemsets
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)

# Print the association rules
print(rules)
```

```python
te = TransactionEncoder()

te_df = te.fit(df).transform(df)

df = pd.DataFrame(te_df,columns=te.columns_)
df = df.astype(int)
df
```
